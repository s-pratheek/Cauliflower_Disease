# Importing Libraries

```python
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_size=256
bat_size=32
EPOCHS=50
```

```python
dataset=tf.keras.preprocessing.image_dataset_from_directory('Cauliflower_Disease_Original_Dataset',
    shuffle=True,
    image_size=(img_size,img_size),
    batch_size=bat_size)
```

```python
class_names=dataset.class_names
class_names
```

```python
len(dataset)
```

# Data Visualtion:

```python
# Printing an image

for image_batch, label_batch in dataset.take(1):
    plt.imshow(image_batch[0].numpy().astype('uint8'))
    plt.axis('off')
    plt.title(class_names[label_batch[0]])
```

```python
# Printing an image

plt.figure(figsize=(10,10))# setting the general display size on the screen
for image_batch, label_batch in dataset.take(1):
    for i in range(12):
        ax=plt.subplot(3,4,i+1)
        # subplot should be used, when displaying multiple images
        plt.imshow(image_batch[i].numpy().astype('uint8'))
        plt.axis('off')
        plt.title(class_names[label_batch[i]])
```

# Splitting the dataset into 'Train' and 'Test'

```python


def get_dataset_partitions_tf(ds, train_split=.8, test_split=.1, val_split=.1,shuffle=True, shuffle_size=10000):
    ds_size=len(ds)
    if shuffle:
        ds=ds.shuffle(shuffle_size,seed=12)
        
    train_ds=ds.take(int(ds_size*train_split))
    test_ds_1=ds.skip(int(ds_size*train_split))
    val_ds=test_ds_1.take(int(val_split*ds_size))
    test_ds=test_ds_1.skip(int(val_split*ds_size))
    
    
    return train_ds, val_ds, test_ds
```

```python
train_ds,val_ds,test_ds=get_dataset_partitions_tf(dataset)
```
