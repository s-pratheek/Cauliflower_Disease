# Importing Libraries

```python
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_size=256
bat_size=32
EPOCHS=50
```

```python
dataset=tf.keras.preprocessing.image_dataset_from_directory('Cauliflower_Disease_Original_Dataset',
    shuffle=True,
    image_size=(img_size,img_size),
    batch_size=bat_size)
```

```python
class_names=dataset.class_names
class_names
```

```python
len(dataset)
```

# Data Visualtion:

```python
# Printing an image

for image_batch, label_batch in dataset.take(1):
    plt.imshow(image_batch[0].numpy().astype('uint8'))
    plt.axis('off')
    plt.title(class_names[label_batch[0]])
```

```python
# Printing an image

plt.figure(figsize=(10,10))# setting the general display size on the screen
for image_batch, label_batch in dataset.take(1):
    for i in range(12):
        ax=plt.subplot(3,4,i+1)
        # subplot should be used, when displaying multiple images
        plt.imshow(image_batch[i].numpy().astype('uint8'))
        plt.axis('off')
        plt.title(class_names[label_batch[i]])
```
